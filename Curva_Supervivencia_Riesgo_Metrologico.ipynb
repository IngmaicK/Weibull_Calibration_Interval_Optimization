```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros validados en la sección 5.2
beta = 1.57392
eta = 726.378
t_propuesto = 110.05
t_actual = 365.0
```

```python
def reliability(t, beta, eta):
    return np.exp(-(t / eta)**beta)

t_espacio = np.linspace(0, 1000, 500)
r_valores = reliability(t_espacio, beta, eta)
r_actual = reliability(t_actual, beta, eta)
```

```python
# Si quieres que la gráfica sea un poco más interactiva en VS Code
# puedes instalar 'ipympl' y usar %matplotlib widget, 
# pero por ahora el estándar funciona perfecto:
%matplotlib inline 

plt.figure(figsize=(10, 6))
plt.plot(t_espacio, r_valores, label='Modelo Weibull (MLE)', color='#1f77b4', lw=2.5)
plt.axhline(y=0.95, color='green', linestyle='--', alpha=0.5)
plt.axvline(x=t_propuesto, color='green', linestyle='--', label=f'Optimizado: {t_propuesto} días')
plt.axvline(x=t_actual, color='red', linestyle=':', label=f'Actual: 365 días (R={r_actual*100:.1f}%)')

plt.fill_between(t_espacio, r_valores, alpha=0.1, color='#1f77b4') # Sombreado bajo la curva
plt.title('Curva de Supervivencia y Riesgo Metrológico')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
