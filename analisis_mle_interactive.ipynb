```python
import plotly.graph_objects as go
import numpy as np

# 1. Parámetros del modelo (Los que ya validamos)
beta = 1.57392
eta = 726.378
t_propuesto = 110.05
t_actual = 365.0

# 2. Generar datos
t_espacio = np.linspace(0, 1000, 1000) # Vector de 0 a 1000 días
r_valores = np.exp(-(t_espacio / eta)**beta)

# 3. Crear la figura interactiva
fig = go.Figure()

# Agregar la curva de confiabilidad
fig.add_trace(go.Scatter(
    x=t_espacio, y=r_valores,
    mode='lines',
    name='Modelo Weibull (MLE)',
    line=dict(color='blue', width=3),
    hovertemplate='Día: %{x:.1f}<br>Confiabilidad: %{y:.4f}<extra></extra>'
))

# Agregar punto del Intervalo Propuesto (R=95%)
fig.add_trace(go.Scatter(
    x=[t_propuesto], y=[np.exp(-(t_propuesto / eta)**beta)],
    mode='markers+text',
    name='Propuesta (R=95%)',
    marker=dict(color='green', size=12, symbol='diamond'),
    text=[f"Propuesto: {t_propuesto} días"],
    textposition="top center"
))

# Agregar punto del Intervalo Actual (R=~71%)
r_actual = np.exp(-(t_actual / eta)**beta)
fig.add_trace(go.Scatter(
    x=[t_actual], y=[r_actual],
    mode='markers+text',
    name='Actual (R=71%)',
    marker=dict(color='red', size=12, symbol='x'),
    text=[f"Actual: 365 días (R={r_actual*100:.1f}%)"],
    textposition="bottom center"
))

# Configuración del diseño (Layout) para la Tesis
fig.update_layout(
    title='<b>Figura 5.2: Análisis Interactivo de Riesgo Metrológico</b>',
    xaxis_title='Tiempo de Operación (Días)',
    yaxis_title='Confiabilidad R(t)',
    yaxis=dict(range=[0, 1.05], tickformat='.2%'),
    xaxis=dict(gridcolor='lightgray'),
    plot_bgcolor='white',
    hovermode='x unified',
    legend=dict(yanchor="top", y=0.99, xanchor="right", x=0.99)
)

# Mostrar la gráfica
fig.show()
```
